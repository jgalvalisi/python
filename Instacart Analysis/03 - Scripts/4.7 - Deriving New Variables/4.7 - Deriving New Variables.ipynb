## 03 - If-Statements with User-Defined Functions

```python
# Create a subset

df = ords_prods_merged[:1000000]
```

```python
df.shape
```

```python
# Define a function

def price_label(row):

    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High-range product'
    else: return 'Not enough data'
```

```python
# Apply the function

df['price_range'] = df.apply(price_label, axis = 1)
```

```python
# Check the values in the new column

df['price_range'].value_counts(dropna = False)
```

```python
# Check what the most expensive product within the subset is

df['prices'].max()
```

## 04 - If-Statements with the loc() Function

### a) Subset

```python
# Create the functions with .loc

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# Check the values in the new column

df['price_range_loc'].value_counts(dropna = False)
```

### b) Entire data frame

```python
# Create the three functions with .loc

ords_prods_merged.loc[ords_prods_merged['prices'] > 15, 'price_range'] = 'High-range product'
```

```python
ords_prods_merged.loc[(ords_prods_merged['prices']  > 5) & (ords_prods_merged['prices'] <= 15), 'price_range'] = 'Mid-range product' 
```

```python
ords_prods_merged.loc[ords_prods_merged['prices'] <= 5, 'price_range'] = 'Low-range product'
```

```python
# Check the values in the new column

ords_prods_merged['price_range'].value_counts(dropna = False)
```

## 05 - If-Statements with For-Loops

```python
# Check the frequency in the orders_day_of_the_week column

ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Create the new busiest_day column

result = []

for x in ords_prods_merged["orders_day_of_week"]:
  if x == 0:
    result.append("Busiest day")
  elif x == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# Print the result

result
```

```python
# Create a new column within the ords_prods_merge dataframe and set it equal to 'result'

ords_prods_merged['busiest_day'] = result
```

```python
# Check the frequency in the new busiest_day column

ords_prods_merged['busiest_day'].value_counts(dropna = False)
```
